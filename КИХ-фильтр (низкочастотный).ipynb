```python
import numpy as np
```

```python
from scipy.signal import butter, filtfilt
```

```python
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('D:/Computer/OneDrive/Predmeti/Цифровая обработка сигнала/res.txt', delimiter=',')
```

```python
# Получаем данные их первого столбца таблицы (ускорение, торможение)
col1 = data[:, 0]
signal = col1[:60000]
lenFiltr = len(signal)
print("Количество строк в списке:", lenFiltr)
```

```python
# Задаем параметры фильтра
nyquist_freq = 0.5 * 1000  # частота Найквиста
cutoff_freq = 2           # частота среза
filter_order = 4          # порядок фильтра
```

```python
# Расчитываем коэффициенты КИХ-фильтра
cutoff = cutoff_freq / nyquist_freq
b, a = butter(filter_order, cutoff, btype='low')
```

```python
# Фильтруем сигнал
filtered_signal = filtfilt(b, a, signal)
```

```python
# Отображаем графики исходного и отфильтрованного сигналов

plt.plot(signal, label='Original signal')
plt.plot(filtered_signal, label='Filtered signal')
plt.legend()
plt.show()
```
